```python
#Q1 print numbers from 1 from 10 usig loop 
for i in range(11):
    print(i)
```

```python
#Q2 print from 10 to 1
for i in range(10,0,-1):
    print(i)
```

```python
#Q3 print even no from 1,50
for i in range(0,50,2):
    print(i)
```

```python
#Q4 print the table of 7
for i in range (7,71,7):
    print(i)
```

```python
#Q5 print sum of n natural numbers
n = 10
res = 0
for i in range(1,n+1,1):
    res += i
print(res)
```

```python
#Q6 print factorial of any number
n = 5
fact = 1

for i in range(1,n+1):
    fact = i*fact
print(fact)
```

```python
#Q7 Count number of digits in python
n = 1023230
count = 0
while(n):
    count+=1
    n = n//10
print(count)
```
