#Anatomy of Scope
#InstancemClas and Static Members
#          Instance Member      Class Membber       Static Member
Decorator   None                        @classMethod                    @statiocmethod
First Arg   self(referes to object)     cls(refers to argument)     No Mandatory argument
Access      Instance                    Class level data            No access to state

```python
class Employee:
    #class variable
    company_name = 'TechCorp'

    def __init__(self,name):
        self.name = name #instance variable

    def show_name(self):
        print(f'Company Name:(self.company_name) Employees:(self.name)')
    
    @classmethod
    def change_company(cls,new_name):
        cls.company_name = new_name

    @staticmethod
    def is_work_day(day):
        return day not in ["Saturday","Sunday"]
    
```

```python
#usage 
emp =Employee('John') #instance or object
print(Employee.company_name)
print(emp.name)
print(emp.is_work_day("Monday"))#called by object
print(Employee.is_work_day("Monday"))#called by class
```

```python
class BankAccount:
    bank_name = "Global Bank"
    def __init__(self,owner,balance):
        self.owner = owner
        self.balance = balance
    
    def deposit(self,amount):
        self.balance+= amount
        print(f'(amount) deposited,New balance: (self.balance)')
    
    def withdraw(self,amount):
        self.balance -= amount
        print(f'(amount) withdrawn, remaining balance: (self.balance)')
```

```python
class SavingAccount(BankAccount):
    intrest_rate = 0.05

    def add_intrest(self):
        intrest = self.balance*self.intrest_rate
        self.balance += intrest
        print(f"Intrest added: {intrest}")
```

```python
class CurrentAccount(BankAccount):
    overdraft_limit = 1000

    def withdraw(self,amount):
        if (self.balance - amount >= self.overdraft_limit):
            self.balance -= amount
            print(f'{amount} Withdrawn, Balance: {self.balance}')
        else:
            print("Overdraft limit exceeded")
```

```python
sa = SavingAccount('James',50000)
sa.add_intrest()
ca = CurrentAccount('James',50000)
ca.withdraw(1000)
```
