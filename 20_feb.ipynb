POLYMORPHISH : One name Many functionalities

Method overridding in python

```python
class PaymentMethod:
    def pay(self,amount):
        raise NotImplementedError("Subclasses must implement this method")
```

```python
class CreditCardPayment(PaymentMethod):
    def pay(self,amount):
        return f'Preocessing Credit card payment of RS. {amount}'
class PayPalPayment(PaymentMethod):
    def pay(self,amount):
        return f'Preocessing PayPal payment of RS. {amount}'
class CryptoPayment(PaymentMethod):
    def pay(self,amount):
        return f'Preocessing Crypto payment of RS. {amount}'
```

```python
def process_payment(payment_method,amount):
    print(payment_method.pay(amount))
```

```python
payments = [CreditCardPayment(),PayPalPayment(),CryptoPayment()]
for pay in payments:
    process_payment(pay,500)
```

#Pillar-4

ABSTRACTION : The Contract of Hiding Complexitya and Showing Features

```python
from abc import ABC,abstractmethod
```

```python
class PaymentMethod(ABC):
    @abstractmethod
    def pay(self,amount):
        pass
```

```python
class CreditCardPayment(PaymentMethod):
    def pay(self,amount):
        return f'Preocessing CreditCard payment of RS. {amount}'
class PayPalPayment(PaymentMethod):
    def pay(self,amount):
        return f'Preocessing PayPal payment of RS. {amount}'
class CryptoPayment(PaymentMethod):
    def pay(self,amount):
        return f'Preocessing Crypto payment of RS. {amount}'
```

```python
obj = CreditCardPayment()
print(obj.pay(500))
```

#Static Vs Dynamic Binding

In Pyhton everting is Dynamic Binding not like C++

The Diamond problem in handled by  `D.mro() (Method Resulution Model)

        A
       / \
      B   C
       \ /
        D

Python searches methods in this order.

So when super() is called:

D → calls B

B → calls C

C → calls A

A → stops

Without D.mro() A-> show might be called twice.

No duplication. No confusion.

```python
class A:
    def show(self):
        print("Inside class A")

class B(A):
    def show(self):
        print("Inside class B")
        super().show()

class C(A):
    def show(self):
        print("Inside class C")
        super().show()

class D(B, C):
    def show(self):
        print("Inside class D")
        super().show()


obj = D()
obj.show()
```
