```python
#Q1 Apply Binary search in nearly sorted array
def search_nearly_sorted(arr,k):
    low,high = 0,len(arr) -1
    while(low <= high):
        mid = (low+high)//2
        if(arr[mid] == k):
            return mid
        elif (mid-1 >= low and arr[mid -1] == k):
            return mid-1
        elif(mid+1 <= high and arr[mid+1] == k):
            return mid+1
        if(k< arr[mid]):
            high = mid-2
        else:
            low = mid+2

arr = [10, 3, 40, 20, 50, 80, 70]
k =  40
print(search_nearly_sorted(arr,k))
```
