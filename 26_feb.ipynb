```python
def linear_search(arr,k):
    for i in range(len(arr)):
        if(arr[i] == k):
            return print("present at index : ", i)
    return print("number not found");
```

```python
def Binary_search(arr,k):
    start = 0
    end = len(arr)-1
    while(start <= end):
        mid = start + (end - start)//2;
        if(arr[mid] == k):
            return print("present at index : ", mid)
        elif(arr[mid] < k):
            start = mid +1
        else:
            end = mid - 1
    return
```
